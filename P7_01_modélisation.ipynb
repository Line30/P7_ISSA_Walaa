```python
def convert_types(df, print_info = False):
    
    original_memory = df.memory_usage().sum()
    
    # Iterate through each column
    for c in df:
        
        # Convert ids and booleans to integers
        if ('SK_ID' in c):
            df[c] = df[c].fillna(0).astype(np.int32)
            
        # Convert objects to category
        elif (df[c].dtype == 'object') and (df[c].nunique() < df.shape[0]):
            df[c] = df[c].astype('category')
        
        # Booleans mapped to integers
        elif list(df[c].unique()) == [1, 0]:
            df[c] = df[c].astype(bool)
        
        # Float64 to float32
        elif df[c].dtype == float:
            df[c] = df[c].astype(np.float32)
            
        # Int64 to int32
        elif df[c].dtype == int:
            df[c] = df[c].astype(np.int32)
        
    new_memory = df.memory_usage().sum()
    
    if print_info:
        print(f'Original Memory Usage: {round(original_memory / 1e9, 2)} gb.')
        print(f'New Memory Usage: {round(new_memory / 1e9, 2)} gb.')
        
    return df
```

```python
def clean_dataset(df):
    assert isinstance(df, pd.DataFrame), "df needs to be a pd.DataFrame"
    df.dropna(inplace=True)
    indices_to_keep = ~df.isin([np.nan, np.inf, -np.inf]).any(1)
    return df[indices_to_keep].astype(np.float64)
```

```python
def pca_reduction(feature_vector,n):
    """PCA reduction"""
    print("Dimensions dataset avant réduction PCA : ", feature_vector.shape)
    pca = decomposition.PCA(n_components=n)
    feat_pca= pca.fit_transform(feature_vector)
    print("Dimensions dataset après réduction PCA : ", feat_pca.shape)
    
    """Scree plot"""
    scree = pca.explained_variance_ratio_*100
    #plt.bar(np.arange(len(scree))+1, scree)
    plt.plot(np.arange(len(scree))+1, scree.cumsum(),c="red",marker='o')
    plt.xlabel("Rang de l'axe d'inertie")
    plt.ylabel("Pourcentage d'inertie")
    plt.title("Eboulis des valeurs propres")
    plt.show(block=False)
    
    return(pca, feat_pca)
```

```python
def tsne_reduction(data_train, data_labels,n_perplexity, n_random_state):
    """tsne reduction"""
    tsne = manifold.TSNE(n_components=2, perplexity=n_perplexity, 
                     n_iter=2000, init='pca', random_state=n_random_state) # for projection in a plane, initialisation by a PCA
    X_tsne = tsne.fit_transform(data_train)

    df_tsne = pd.DataFrame(X_tsne[:,0:2], columns=['tsne1', 'tsne2'])
    df_tsne["Target"] = data_labels
    
    """tsne visualisation"""
    plt.figure(figsize=(8,8))
    sns.scatterplot(x="tsne1", y="tsne2", hue="Target", data=df_tsne, legend="brief")

    plt.title('TSNE selon les vraies classes', fontsize = 20, pad = 35, fontweight = 'bold')
    plt.xlabel('tsne1', fontsize = 16, fontweight = 'bold')
    plt.ylabel('tsne2', fontsize = 16, fontweight = 'bold')
    plt.legend(prop={'size': 14}, bbox_to_anchor=(1,0), loc='lower left') 

    plt.show()
```

## Définition de la fonction coût 

Définition de La "fonction coût" :

Erreur de prédiction :
* FP (False Positive) : les cas où la prédiction est positive, mais où la valeur réelle est négative.

   ----------> Perte d'opportunité si le crédit client est refusé à tort, alors qu'il aurait été en mesure d'être remboursé.
   


* FN (False Negative) : les cas où la prédiction est négative, mais où la valeur réelle est positive.

   ----------> Perte réelle si le crédit client accepté se transforme en défaut de paiement.

Ainsi, les pertes d'un crédit en raison d'une mauvaise classification dépendront des probabilités Faux Positifs et Faux Négatifs.

L'idée est d'éviter les clients avec un fort risque de défaut. Il est donc nécessaire de pénaliser les FP et FN cités précédemment. Pour réduire ce rique de perte financière, il faut maximiser les deux critères Recall et Precision. 

* Précision :

<img src="images/precision.png" width="200"/>

* Recall :

<img src="images/recall.png" width="200"/>

 

La fonction qui optimise Precision et Recall est la fonction F beta score :

<img src="images/F_beta.png" width="600"/>
 

Avec Beta le coefficient d'importance relative au critère Precision par rapport au recall :

* Pour beta > 1, on accorde plus d’importance au recall (autrement dit aux faux négatifs).
* Pour beta < 1, on accorde plus d’importance à la precision (autrement dit aux faux positifs).
* Pour beta = 1, on retrouve le F1-score, qui accorde autant d’importance à la precision qu’au recall :

<img src="images/f1_score.JPG" width="200"/>

```python
def learning_algorithms (Algorithms,scoring ,X_train, X_test, y_train, y_test, à_predir, figure_title):
    algorithmes=[]  
    y_predit=[]
    pred_proba_values = {}
    best_estimator_alg=[]
    ARI=[]
    time_alg = []
    f3_score = []
    AUC = []
    Accuracy= []
    Precision = []
    Recall = []
    F1 = []
    
    x_train, x_test= standarisation (X_train, X_test)
    
    for algorithm_id, algorithm in Algorithms.items():
        print('-----------------------------------------------------------------------------------------------------------')
        print('\nAlgorithme :', algorithm_id )
        print('-----------------------------------------------------------------------------------------------------------')
        estimator=algorithm['estimator']
        param_grid=algorithm['parameters_grid']
        print('Parameters grid :', param_grid )
        
        start = time.time()
        Randomized_search=RandomizedSearchCV(estimator, param_grid, scoring=scoring, cv=5, verbose=2, n_jobs=-1) #refit="fbeta"
        grid_result = Randomized_search.fit(x_train, y_train)
        print('Time :',time.time()-start )
        print('Best score :',grid_result.best_score_)
        print('Best parameters :',grid_result.best_params_)
        
        best_estimator=grid_result.best_estimator_
        best_estimator_alg.append(best_estimator)
        time_alg.append(time.time()-start)
        
        print('_________________________________________')
        print('Train Accuracy : %0.3f' %best_estimator.score(x_train,y_train))
        print('test Accuracy :%0.3f' %best_estimator.score(x_test,y_test))
        print('_________________________________________')
        
        y_pred=best_estimator.predict(x_test) 
        y_predit.append(y_pred)
        y_pred_proba=best_estimator.predict_proba(x_test)
        pred_proba_values[algorithm_id]= y_pred_proba
        
        """calcul ARI de similarité catégories"""
        ARI_ = metrics.adjusted_rand_score(y_test, y_pred)
        f3_score_ = fbeta_score(y_test, y_pred, beta=3)
        AUC_= roc_auc_score(y_test, y_pred_proba[:,1])
        Accuracy_=  accuracy_score(y_test, y_pred)
        Precision_= precision_score(y_test, y_pred)
        Recall_= recall_score(y_test, y_pred)
        F1_= f1_score(y_test, y_pred)
        
        print("fBeta score (beta=3): %0.3f " % f3_score_)
        print("AUC: %0.3f " % AUC_)
        print("Accuracy: %0.3f "% Accuracy_)
        print("Precision: %0.3f "% Precision_)
        print("Recall: %0.3f "% Recall_)
        print("F1: %0.3f "% F1_)
        print("ARI: %0.3f "% ARI_)
        print("V-measure: %0.3f "% metrics.v_measure_score(y_test, y_pred))
        print("Adjusted Mutual Information: %0.3f "% metrics.adjusted_mutual_info_score(y_test, y_pred))
        algorithmes.append(algorithm_id)
        ARI.append(ARI_)
        f3_score.append(f3_score_)
        AUC.append(AUC_)
        Accuracy.append(Accuracy_)
        Precision.append(Precision_)
        Recall.append(Recall_)
        F1.append(F1_)
                    
        print('-----------------------------------------------------------------------------------------------------------')
        print('-----------------------------------------------------------------------------------------------------------')
    
    pred_values = pd.DataFrame(list(zip(algorithmes,y_predit)), columns =['Algorithmes','y_pred'])
    #pred_proba_values = pd.DataFrame(list(zip(algorithmes,y_pred_proba)), columns =['Algorithmes','y_pred_proba'])
    
    print('-----------------------------------------------------------------------------------------------------------')
    print('\nComparaison des algorithmes:' )
    print('-----------------------------------------------------------------------------------------------------------')
    
    comparaison = pd.DataFrame(list(zip(algorithmes,time_alg, ARI, f3_score, AUC, Accuracy, Precision, Recall, F1 )), 
                               columns =['Algorithmes','Time', 'ARI','f3_score', 'AUC', 'Accuracy', 'Precision', 'Recall', 'F1'])
    comparaison=comparaison.sort_values(by='f3_score', ascending=False)
    comparaison=pd.pivot_table(comparaison, values=['Time', 'ARI','f3_score', 'AUC', 'Accuracy', 'Precision', 'Recall', 'F1'], 
                               columns=['Algorithmes'], aggfunc=np.sum)
    print(comparaison)
    print('-----------------------------------------------------------------------------------------------------------')
    
    comparaison.loc[['f3_score', 'AUC', 'Accuracy', 'Precision', 'Recall', 'F1'], :].plot.bar(rot=0, 
                                                                                              color = ['darkseagreen', '#5F9EA0', 'steelblue',  'indianred'], 
                                                                                              figsize=(8,5))
    plt.legend(bbox_to_anchor=(1,0), loc='lower left')
    plt.xlabel('Mesures de précision')
    plt.title('Comparaison des differents algorithmes')
    plt.savefig("visualization/erreur2_"+ str(figure_title)+".png", bbox_inches='tight')
    plt.show()
    
    print('-----------------------------------------------------------------------------------------------------------')
    print('-----------------------------------------------------------------------------------------------------------')
    print('-----------------------------------------------------------------------------------------------------------')
    
    best_estimators = pd.DataFrame(list(zip(algorithmes,best_estimator_alg)), columns =['Algorithmes','Meilleur estimateur'])
    
    return pred_values, pred_proba_values, comparaison, best_estimators
```

```python
def cf_matrix_roc_auc(y_true, y_pred, y_pred_proba, model):
    '''This function will make a plot of an sklearn Confusion Matrix using a Seaborn heatmap visualization + ROC Curve.'''
    
    print('Algorithme :', model )
    print('------------------------------------------')
    """performance metrics """
    print ('Accuracy Score is',accuracy_score(y_true, y_pred))
    print ('Classification Report : ')
    print (classification_report(y_true, y_pred))
    print('AUC-ROC :',roc_auc_score(y_true, y_pred, multi_class='ovo'))
    auc = roc_auc_score(y_true, y_pred_proba[:,1])
    print('AUC :',auc)
    #print('LOGLOSS Value is',log_loss(y_true, y_pred))
    
    fig = plt.figure(figsize=(20,15))
    
    """ heatmap (Target - prediction) """
    plt.subplot(221)
    cf_matrix = confusion_matrix(y_true, y_pred)
    group_names = ['True Neg', 'False Pos', 'False Neg', 'True Pos']
    group_counts = ["{0:0.0f}".format(value) for value in cf_matrix.flatten()]
    group_percentages = ["{0:.2%}".format(value) for value in cf_matrix.flatten()/np.sum(cf_matrix)]
  
    labels = [f"{v1}\n{v2}\n{v3}" for v1, v2, v3 in zip(group_names,group_counts,group_percentages)]
    labels = np.asarray(labels).reshape(2,2)
    ax = sns.heatmap(cf_matrix, annot=labels, fmt="", cmap='Blues')
    bottom, top = ax.get_ylim()
    ax.set_ylim(bottom + 0.5, top - 0.5)
    ax.set(xlabel='Prediction', ylabel='Target', title='cross table')

    plt.subplot(222)
    fpr,tpr,_ = roc_curve(y_true, y_pred_proba[:,1])
    plt.plot(fpr, tpr, color='orange', linewidth=5, label='AUC = %0.4f' %auc)
    plt.plot([0, 1], [0, 1], color='darkblue', linestyle='--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.legend()
    plt.show()
```

La classe 0 a une précision beaucoup plus élevée que la classe 1 (+83%), mais la classe 1 a un rappel beaucoup plus élevé que la classe 0 (+29%).


Le rapport de classification porte sur les mesures clés d’un problème de classification.

Vous aurez de la précision, du rappel, un score f1 et un support pour chaque classe que vous essayez de trouver.

* Le rappel signifie « combien de cette classe vous trouvez sur le nombre entier d’éléments de cette classe »

* La précision sera « combien sont correctement classés dans cette classe »

* La partition f1 est la moyenne harmonique entre précision et rappel

* Le support est le nombre d’occurrence de la classe donnée dans le jeu de données (70687 de classe 0 et 6190 de classe 1, ce qui est un jeu de données non équilibré.)

La précision et le rappel sont très utilisés pour les ensembles de données déséquilibrés, car dans un ensemble de données très déséquilibré, une précision de 99% peut être dénuée de sens.
